### Task 1: Write a well-documented function to generate random numbers from scratch without using any libraries in python

**Solution**

Here i am using Linear Congruential Generator algorithm for generating Uniform Random numbers.

>> Random number Formula = (a * S + b) % M


```python
def random_num_generator(N,S):
    # N = size or numbers of random nmuber
    # S = Seed
    
    a = 7**5
    M = 2**31-1
    
    def ran_gen(S):
        return (a*S)%M
    
    U = []
    
    for i in range(N):
        S = ran_gen(S)
        U +=[S/M]
        
    return U

print(random_num_generator(10,123456789))
```

### Task 3: predict the value of pi using the generated number

```python
def random_num_generator(N,S):
    # N = size or numbers of random nmuber
    # S = Seed
    
    a = 7**5
    M = 2**31-1
    
    def ran_gen(S):
        return (a*S)%M
    
    U = []
    
    for i in range(N):
        S = ran_gen(S)
        U +=[S/M]
        
    return U

def pi_estimator(n):
    
    pointsInCircle = 0
    
    X = random_num_generator(n,374441)
    Y = random_num_generator(n,674441)
    
    for i in range(n):
        
        distance = (X[i]*X[i]+Y[i]*Y[i])**0.5
        
        if distance < 1:
            pointsInCircle +=1
        
    return (4*pointsInCircle/n)

print("Pi value :",pi_estimator(1000000))
```
